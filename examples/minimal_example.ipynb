# Minimal StaGPS Example

This notebook runs the packaged algorithm on a tiny synthetic dataset.

```python
import matplotlib.pyplot as plt

from stagps import StaGPS, StaGPSConfig, simulate_toy
```

```python
data = simulate_toy(seed=7)
model = StaGPS(StaGPSConfig(
    candidate_genes=120,
    threshold=0.30,
    patch_cut=1.0,
    min_genes=4,
    max_modules=6,
))
result = model.fit_predict(data)

print(f"cells: {data['X'].shape[0]}")
print(f"genes: {data['X'].shape[1]}")
print(f"stable modules: {len(result['stable_modules'])}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3.5), constrained_layout=True)

truth = data["state"]
ax[0].scatter(data["coords"][:, 0], data["coords"][:, 1], c=truth, s=12, cmap="tab10", linewidths=0)
ax[0].set_title("toy truth")

sc = ax[1].scatter(data["coords"][:, 0], data["coords"][:, 1], c=result["steady_score"], s=12, cmap="magma", linewidths=0)
ax[1].set_title("discovered steady score")
fig.colorbar(sc, ax=ax[1], fraction=0.046, pad=0.02)

for a in ax:
    a.set_xticks([])
    a.set_yticks([])
```

```python
for i, module in enumerate(result["stable_modules"], 1):
    names = data["genes"][module["ranked_genes"][:8]]
    print(f"M{i}: patch={module['patch']:.2f}, genes={', '.join(names)}")
```
